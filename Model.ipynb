```python
new_df = df.copy()
new_df.drop('Poster_Link', axis=1, inplace=True)
new_df.drop('Overview', axis=1, inplace=True)
new_df['No_of_Votes'] = new_df['No_of_Votes'].astype(float)
new_df['Gross'] = new_df['Gross'].str.replace(",", "").astype(float)
new_df['Released_Year'] = pd.to_numeric(new_df['Released_Year'], errors='coerce')
new_df['Runtime'] = new_df['Runtime'].str.replace(' min', '').astype(int)
```

```python
new_df['Gross'] = new_df['Gross'].fillna(new_df['Gross'].median())
new_df['Meta_score'] = new_df['Meta_score'].fillna(new_df['Meta_score'].median())
```

```python
new_df['Released_Year'].values, new_df['Runtime'].info()
```

```python
# Used multilabelbinarizer to create athe same effect as onehotencoder but for more than two classes/columns [custom onehot encoding]

from sklearn.preprocessing import MultiLabelBinarizer
from sklearn.preprocessing import OneHotEncoder
new_df['Genre_List'] = new_df['Genre'].fillna('').str.split(',\s*')
new_df['Genre_List'].value_counts()
new_df.index
```

```python

mlb = MultiLabelBinarizer()
onehot = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_cert = onehot.fit_transform(new_df['Certificate'].values.reshape(-1,1)) #Passing only one feature so converting it into appropriate shape required for OneHotEncoder to work
encoded_cert_df = pd.DataFrame(encoded_cert, columns=onehot.get_feature_names_out(['Certificate']), index= new_df.index)

genre_encoded = mlb.fit_transform(new_df['Genre_List'])

genre_df = pd.DataFrame(genre_encoded, columns=[f'genre_{g}' for g in mlb.classes_])
genre_df.index = new_df.index
new_df = pd.concat([new_df, encoded_cert_df, genre_df], axis=1)
new_df.drop(columns=['Genre', 'Genre_List', 'Certificate'], inplace=True)
```

```python
new_df.isnull().sum()
```

```python
new_df.info()
```

```python
# Lets train our recommendation system using this dataset 
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics.pairwise import cosine_similarity

genre_cols = [col for col in new_df.columns if col.startswith("genre_")]
certificate_cols = [col for col in new_df.columns if col.startswith('Certificate_')]

numeric_cols = ['Meta_score','No_of_Votes','Gross','Runtime','Released_Year']
categorical_cols = ['Director', 'Star1', 'Star2', 'Star3', 'Star4']

scaler = MinMaxScaler()
new_df[numeric_cols] = scaler.fit_transform(new_df[numeric_cols].fillna(0)) # One na in released year

onehot = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
categorical_mtx = onehot.fit_transform(new_df[categorical_cols])
categorical_df = pd.DataFrame(categorical_mtx, columns=onehot.get_feature_names_out(categorical_cols), index = new_df.index)

feature_mtx = pd.concat(
    [
        new_df[genre_cols],
        new_df[certificate_cols],
        new_df[numeric_cols],
        categorical_df,
    ],
    axis = 1
)

cosine_sim = cosine_similarity(feature_mtx, feature_mtx)


def recommend_by_title(title, top_rec = 5):
    if title not in new_df['Series_Title'].values:
        return f"Movie '{title}' not found in database."
    
    index = new_df[new_df['Series_Title'] == title].index[0]
    sim_score = list(enumerate(cosine_sim[index]))
    sim_score = sorted(sim_score, key=lambda x: x[1], reverse= True)

    top_recommendations = [i for i, score in sim_score[1:top_rec+1]]

    return new_df.iloc[top_recommendations][['Series_Title', 'Director', 'Star1', 'IMDB_Rating']]

def recommend_by_genre(genre, top_rec = 5):
    genre_col = f"genre_{genre}"
    if genre_col not in new_df.columns:
        return f"Genre '{genre}' not found in database"
    
    genre_movies = new_df[new_df[genre_col] == 1]
    if genre_movies.empty:
        return f"No movies found for genre '{genre}"
    
    genre_movies = genre_movies.copy()
    genre_movies['score'] = (genre_movies['IMDB_Rating']*0.7) + (genre_movies["No_of_Votes"]*0.3) # Score based more towards IMDB rating rather than no of votes (people) 
    top_movies_rec = genre_movies.sort_values('score', ascending=False).head(top_rec)

    return top_movies_rec[['Series_Title', 'IMDB_Rating', 'Director', 'Star1']]

s = input("Enter the title of movie you have watched: ")
print(f"\nMovies similar to '{s}': ")
print(recommend_by_title(s, top_rec=5))

g = input("Enter the genre of movie you want to watch: ")
print(f"\nTop {g} Movies: ")
print(recommend_by_genre(g, top_rec=5))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

df_1 = df.copy()
df_1.drop('Poster_Link', axis=1, inplace=True)
df_1['No_of_Votes'] = df_1['No_of_Votes'].astype(float)
df_1['Gross'] = df_1['Gross'].str.replace(",", "").astype(float)
df_1['Released_Year'] = pd.to_numeric(df_1['Released_Year'], errors='coerce')
df_1['Runtime'] = df_1['Runtime'].str.replace(' min', '').astype(int)
df_1['Gross'] = df_1['Gross'].fillna(df_1['Gross'].median())
df_1['Meta_score'] = df_1['Meta_score'].fillna(df_1['Meta_score'].median())
```

```python
df_1.info()
```

```python
df_1['Combined_Features'] = (
    df_1['Overview'].fillna("") + ' ' +
    df_1['Director'].fillna("") + ' ' + 
    df_1['Star1'].fillna("") + ' ' + 
    df_1['Star2'].fillna("") + ' ' + 
    df_1['Star3'].fillna("") + ' ' + 
    df_1['Star4'].fillna("") + ' ' + 
    df_1['Genre'].astype(str).str.replace(",", " ")
) # Combined text features for tdilf vectorizer

tfidf = TfidfVectorizer(stop_words='english')
tfidf_mtx = tfidf.fit_transform(df_1['Combined_Features'])

cosine_sim = cosine_similarity(tfidf_mtx, tfidf_mtx)

def recommend_tfidf(title, cosine_sim=cosine_sim): # Here, cosine_sim = cosine_sim says that if cosine_sim has not been provided by user, use the already computed cosine_sim
    if title not in df_1['Series_Title'].values:
        return 'Movie not found in the database'

    idx = df_1[df_1['Series_Title'] == title].index[0]
    similarity_scores = list(enumerate(cosine_sim[idx]))
    similarity_scores = sorted(similarity_scores, key= lambda x: x[1], reverse = True)
    movie_idx = [i for i, score in similarity_scores[1:11]]# top 10 excluding the movie itself, Similarity_scores -> (index, similarity score)
    return df_1.iloc[movie_idx][['Series_Title', 'Director', 'IMDB_Rating', 'Star1']]

s = input("Enter the name of movie you have watched: ")
print(f"Recommended Movies based on '{s}':")
print(recommend_tfidf(s))

# This model usually performs better as Tf-idf Vectorizer converts the text from description and other things into numbers depending on the frequency and creates vector representation
# Then this vector vectorizer can be used for similarities using the cosine similarity function and accurate recommendtions can be made
```
